```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('data2.csv', skiprows=[0], encoding = 'ISO-8859-1')
cols = ['Natural', 'XS', 'S', 'M', 'L', 'Toe', 'Heel', 'Ascending', 'Descending']
```

```python
data.head()
```

```python
def modifyCsv(data):
    t1 = 3
    t2 = 4
    #for young
    for i in cols:
        data[i] = (data[i] + data['Unnamed: ' + str(t1)] + data['Unnamed: ' + str(t2)])/3
        data.drop(['Unnamed: ' + str(t1), 'Unnamed: ' + str(t2)], inplace = True, axis = 1)
        t1 += 3
        t2 += 3
        
    #for adult
    for i in cols:
        data[i + '.1'] = (data[i + '.1'] + data['Unnamed: ' + str(t1)] + data['Unnamed: ' + str(t2)])/3
        data.drop(['Unnamed: ' + str(t1), 'Unnamed: ' + str(t2)], inplace = True, axis = 1)
        t1 += 3
        t2 += 3
        
    
    return data
```

```python
copy_data = data.copy()
copy_data = modifyCsv(copy_data)
copy_data.head()
```

```python
adult = copy_data.copy()
young = copy_data.copy()
```

```python
young.drop([i + '.1' for i in cols], axis=1, inplace=True)
young['Target'] = 'Young'
young.head()
```

```python
for i in cols:
    adult[i] = adult[i + '.1']
    adult.drop([i + '.1'], inplace=True, axis=1)
adult['Target'] = 'Adult'
```

```python
desired_data = pd.concat([young, adult], ignore_index=True)
desired_data.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
```

```python
X = desired_data[cols]
Y = desired_data['Target']
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, random_state = 0)
```

```python
mod = DecisionTreeClassifier().fit(X_train, Y_train)
```

```python
predict = mod.predict(X_test)
```

```python
mod.score(X_test, Y_test)
```
