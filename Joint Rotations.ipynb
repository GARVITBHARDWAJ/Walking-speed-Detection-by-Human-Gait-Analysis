```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv("data1.csv", skiprows=[0], encoding = 'ISO-8859-1')
```

```python
def modify(data):
    col = ["Natural", "XS", "S", "M", "L", "Toe", "Heel", "Ascending", "Descending"]
    m1 = 3
    m2 = 1
    m3 = 1
    d = {}
    r = data.columns[0]
    for i in col:
        s1 = "Unnamed: " + str(m1)
        s2 = "Unnamed: " + str(m1+1)
        data[i] = (data[i] + data[s1] + data[s2])/3
        data.drop([s1, s2], axis = 1, inplace = True)
        m1 += 3
    for i in col:
        a = i + "." + str(m2)
        s1 = "Unnamed: " + str(m1)
        s2 = "Unnamed: " + str(m1+1)
        data[i] = (data[i] + data[s1] + data[s2])/3
        data.drop([s1, s2], axis = 1, inplace = True)
        m1 += 3
        m2 += 1
    for i in data[r].unique():
        d[i]  = m3
        m3 += 1
#     print(d)
    data[r] = data[r].map(d)
    data['% Gait Cycle'] = data['% Gait Cycle'].apply(lambda x : int(x.split('%')[0]))
    return data
```

```python
modify(data.copy()).head()
```

```python
d = modify(data.copy())
young_data = d.copy()
adult_data = d.copy()
col = ["Natural", "XS", "S", "M", "L", "Toe", "Heel", "Ascending", "Descending"]
for i in col:
    young_data.drop(i+".1", axis = 1, inplace = True)
    adult_data[i] = adult_data[i+".1"]
    adult_data.drop(i+".1", axis = 1, inplace = True)
young_data["Target"] = "Young"
adult_data["Target"] = "Adult"
```

```python
data_new = pd.concat([young_data, adult_data], ignore_index=True)
```

```python
data_new.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
```

```python
columns = list(data_new.columns)
columns.remove('Target')
X = data_new[columns]
Y = data_new["Target"]
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, random_state = 0)
```

```python
clf = DecisionTreeClassifier().fit(X_train, Y_train)
```

```python
predictions = clf.predict(X_test)
# predictions
```

```python
clf.score(X_test, Y_test)
```
